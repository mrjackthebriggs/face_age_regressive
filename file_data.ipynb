# For Showing the Grpah data for the file directories

```python
grph_file_dir = "data/train"
grph_test_dir = "data/test"

```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
plt.figure().set_figwidth(20)
plt.figure().set_figheight(20)

plt.style.use('_mpl-gallery')

max_len = len(os.listdir(grph_file_dir)) if len(os.listdir(grph_file_dir)) > len(os.listdir(grph_test_dir)) else len(os.listdir(grph_test_dir))

y1 = list(range(0, max_len))
y2 = list(range(0, max_len))

for file in os.listdir(grph_file_dir):
    file_count = 0
    
    for filf in os.listdir(os.path.join(grph_file_dir,file)):
        file_count += 1
        
    y1[int(file)] = file_count


for file in os.listdir(grph_test_dir):
    file_count = 0
    
    for filf in os.listdir(os.path.join(grph_test_dir,file)):
        file_count += 1
        
    y2[int(file)] = file_count

# make data:
x1 = np.arange(max_len)

width = 1

plt.figure(figsize=(8,3))

# plot
plt.bar(x1, y1, width=width, 
        color='b', label='')
plt.bar([i+0.25*width for i in x1], y2, 
        width=0.7*width, color='r', label='')



plt.xticks([i for i in range(0,len(x1),10)], 
           ['{}'.format(i) for i in range(0,len(y1),10)] )

plt.xlabel("Ages")
plt.ylabel("Number of Imgages")
plt.title("Distribution of Images over Ages (UTKFACE)")

plt.legend(["Training Data", "Test Data"])

plt.show()
```
